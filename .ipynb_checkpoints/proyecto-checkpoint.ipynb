## Ejecución

```python
# Definir las cadenas necesarias
lPA = "Paciente"
lS = "Sexo"
lE = "Edad"
lPe = "Peso"
lA = "Estatura"
lIMC = "IMC"
lDi = "Diabetes"
lH = "Hipertensión"
lDe = "Depresión"
lTs = "Trastorno del sueño"
lTA = "Trastorno de la alimentación"
```

```python
### import colorama
from colorama import Fore, Back, Style
#colorama.init()

num = (input("Dame los números de los pacientes separados por comas"))
numcom = num.count(",")+1
pacientes = num.split(",")

print(50 * chr(0x2015))
print("Sistema XYZ". center(50))
print(50 * chr(0x2015))
print("ID. pacientes (separados por coma) :", num)
print(50 * chr(0x2015))

f = open("personal_data.csv", "r")

PA = []
S = []
E = []
PE = []
A = []
IMC = []
Di = []
H = []
De = []
Ts = []
TA = []

encabezado = f.readline().split(',')

for renglon in f:
    datos = renglon.split(',')
    PA.append(int(datos[0]))
    S.append(datos[1])
    E.append(datos[2])
    PE.append(datos[3])
    A.append(datos[4])
    IMC.append(float(datos[5]))
    Di.append(datos[6])
    H.append(datos[7])
    De.append(datos[8])
    Ts.append(datos[9])
    TA.append(datos[10])

for n in range(0,numcom):
    
    nm=int(pacientes[n])
    nm1=nm-1
    pa = PA[nm1]
    sx = S[nm1]
    est = E[nm1]
    pe = PE[nm1]
    alt = A[nm1]
    ic = IMC[nm1]
    di = Di[nm1]
    hip = H[nm1]
    dep = De[nm1]
    tso = Ts[nm1]
    tal = TA[nm1]
       
    # Formatos para imprimir

    paw=str(pa)
    print(f"{lPA:>32s}: {paw:<18s}")
    print(f"{lS:>32s}: {sx:<18s}")
    print(f"{lE:>32s}: {est:<18s}")
    print(f"{lPe:>32s}: {pe:<18s}")
    print(f"{lA:>32s}: {alt:<18s}")
    if ic >30:
        ics = str(ic)
        print(f"{Fore.RED + lIMC:>37s}: {Fore.RED + ics:<18s}")
    else:
        print(f"{lIMC:>32s}: {ics:<18s}")
    
    if di == "Si":
        print(f"{Fore.RED + lDi:>37s}: {Fore.RED + 'Si' + Style.RESET_ALL :<18s}")
    else:
        print(f"{lDi:>32s}: {di:<18s}")
 
    if hip == "Si":
        print(f"{Fore.RED + lH:>37s}: {Fore.RED + 'Si' + Style.RESET_ALL:<18s}")
    else:
        print(f"{lH:>32s}: {hip:<18s}")

    if dep =="Si":
        print(f"{Fore.RED + lDe:>37s}: {Fore.RED + 'Si' + Style.RESET_ALL:<18s}")
    else:
        print(f"{lDe:>32s}: {dep:<18s}")

    if tso == "Si":
        print(f"{Fore.RED + lTs:>37s}: {Fore.RED + 'Si' + Style.RESET_ALL:<18s}")
    else:    
        print(f"{lTs:>32s}: {tso:<18s}")

    if tal == "Si\n":
        print(f"{Fore.RED + lH:>37s}: {Fore.RED + 'Si' + Style.RESET_ALL:<18s}")
    else:
        print(f"{lTA:>32s}: {'No':<18s}")
    
    print(50 * chr(0x2015))
      
f.close()
```
